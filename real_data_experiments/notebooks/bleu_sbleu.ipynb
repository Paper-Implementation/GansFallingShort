## entropy plot BLEU

```python
data = pd.read_csv('csv/results_entropy_alpha_HBLEU1.csv')
# data
```

```python
temperature = data[data.model.map(lambda x: 'alphat' in x)]
temperature['color'] = temperature.model.map(lambda x: float(x.split('alphat')[1].split('_')[0]))
temperature= temperature[temperature.color  >0 ]
temperature.color = temperature.color ** -1
```

```python
temperature.color
```

```python
cm = plt.cm.get_cmap('Reds')

sc = plt.scatter(-temperature.BLEU.values, \
                 temperature.SBLEU.values, \
                 c=temperature.color.values, vmin=1, vmax=3, s=50, cmap=cm, edgecolors='k')
#plt.colorbar(sc)
plt.xlabel('Negative BLEU-5')
plt.ylabel('SBLEU-5')
plt.axis([-0.6, -0.1, 0.2, 1.05])
plt.colorbar()
plt.savefig('figs/training_temperature_tuning.png')
plt.show()
```
